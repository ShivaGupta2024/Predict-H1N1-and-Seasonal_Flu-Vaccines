H1N1 Vaccine

```python
# Define manual mappings for categorical features
mapping_dict = {
    'age_group': {'18 - 34 Years': 0, '35 - 44 Years': 1, '45 - 54 Years': 2, '55 - 64 Years': 3, '65+ Years': 4},
    'education': {'< 12 Years': 0, '12 Years': 1, 'Some College': 2, 'College Graduate': 3},
    'income_poverty': {'Below Poverty': 0, '<= $75,000, Above Poverty': 1, '> $75,000': 2},
    'race': {'White': 0, 'Black': 1, 'Hispanic': 2, 'Other or Multiple': 3},
    'sex': {'Male': 0, 'Female': 1},
    'marital_status': {'Not Married': 0, 'Married': 1},
    'rent_or_own': {'Rent': 0, 'Own': 1},
    'employment_status': {'Not in Labor Force': 0, 'Unemployed': 1, 'Employed': 2},
    'census_msa': {'Non-MSA': 0, 'MSA, Not Principle City': 1, 'MSA, Principle City': 2}
}

# Apply the mapping, but only if the column exists
for col, mapping in mapping_dict.items():
    if col in df.columns:  # Check if column exists before mapping
        df[col] = df[col].map(mapping)
    else:
        print(f"Warning: Column '{col}' not found in DataFrame. Skipping mapping.")

# Check if the transformation was successful
df.head()
```

```python
# Check unique values in census_msa
print(df['census_msa'].unique())

# Define mapping with a default value for unknowns
census_msa_mapping = {
    'Non-MSA': 0,
    'MSA, Not Principle City': 1,
    'MSA, Principle City': 2
}

# Apply mapping and replace NaN with a default value (-1)
df['census_msa'] = df['census_msa'].map(census_msa_mapping).fillna(df['census_msa'].median())

```

```python
df.isnull().sum()
```

```python
X=df[['h1n1_concern', 'h1n1_knowledge','behavioral_antiviral_meds', 'behavioral_avoidance',
    'behavioral_face_mask', 'behavioral_wash_hands', 'behavioral_large_gatherings', 'behavioral_outside_home',
    'behavioral_touch_face', 'doctor_recc_h1n1', 'doctor_recc_seasonal', 'chronic_med_condition',
    'child_under_6_months', 'health_worker', 'health_insurance', 'opinion_h1n1_vacc_effective',
    'opinion_h1n1_risk', 'opinion_h1n1_sick_from_vacc', 'opinion_seas_vacc_effective', 'opinion_seas_risk',
    'opinion_seas_sick_from_vacc', 'age_group', 'education', 'race', 'sex', 'income_poverty', 'marital_status',
    'rent_or_own', 'employment_status', 'census_msa', 'household_adults', 'household_children'
   ]]

# Y1=np.array(df['h1n1_vaccine']).reshape(-1,1)
Y1=df['h1n1_vaccine']
# Y2=np.array(df['seasonal_vaccine']).reshape(-1,1)
Y2=df['seasonal_vaccine']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y1,test_size=0.30,random_state=42)
```

```python
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
model_h = LogisticRegression()
model_h.fit(x_train,y_train)
```

```python
y_pred=model_h.predict(x_test)
y_pred
```

```python
print('accuracy H1N1 :', accuracy_score(y_test,y_pred)*100)
```

```python
clrep=classification_report(y_pred,y_test)
print('accuracy H1N1 :', accuracy_score(y_test,y_pred)*100)
print("Classification  Seasonal flu:\n",clrep)
```

```python
confusion_matrix=confusion_matrix(y_pred,y_test)
confusion_matrix
```
